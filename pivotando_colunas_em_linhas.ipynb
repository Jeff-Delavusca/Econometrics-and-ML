```python
import pandas as pd
```

```python
caminho = '/content/DadosFinanceiros.xlsx'
```

```python
DadosFinanceiros = pd.read_excel(caminho)
DadosFinanceiros.head(10)
```

```python
type(DadosFinanceiros)
```

```python
DadosFinanceiros.shape
```

```python
df = DadosFinanceiros.copy()
```

```python
df.head(5)
```

```python
df.shape
```

```python
# Identificação das colunas fixas
colunas_fixas = ["Tipo", "Componente"]

# Identificando as todas as colunas que não são fixas que precisam virar linha
colunas_meses = [c for c in df.columns if c not in colunas_fixas]

# Fazendo unpivot (melt)
df_long = df.melt(
    id_vars = colunas_fixas,
    value_vars = colunas_meses,
    var_name = "Data",
    value_name = "Valor"
)
```

```python
df_long.head(5)
```

```python
df_long.shape
```

```python
df_long.info()
```

```python
# Covertendo a coluna Data para o formato de data
df_long["Data"] = pd.to_datetime(df_long["Data"], dayfirst = True)
```

```python
# Ajustando a coluna Valor para o formato americano
df_long["Valor"] = (
    df_long["Valor"]
    .astype(str)
    .str.replace(".", "", regex=False)      # remove pontos como separador de milhar (30.000,00)
    .str.replace(",", ".", regex=False)     # troca vírgula por ponto (30000.00)
)

# Covertendo a coluna Valor para o formato numérico
df_long["Valor"] = pd.to_numeric(df_long["Valor"], errors="coerce")
```

```python
# Exportando a bse ajustada
df_long.to_excel("DadosFinanceirosAjustado.xlsx", index=False)
```
